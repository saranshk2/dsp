```python
fs, data = wavfile.read('sound.wav')
Audio('sound.wav')
```

```python
audio  = data[:,0]
```

```python
audio_dct = dct(audio, norm = 'ortho')
```

```python
print(audio_dct)
len(audio)
```

```python
M = 5000   #number of points to geting deleted 
N = len(audio)
#compact DCT sequence
audio_m_dct = audio_dct[:(N-M)]
print (len(audio_m_dct))
plt.plot((np.arange(len(audio_m_dct)) / fs)*1000, audio_m_dct)
plt.title('Compressed file')
plt.xlabel('Time [ms]')
plt.ylabel('Amplitude')
plt.show()
```

```python
#Insering zero values 
mod_dct = [0]*N
for i in range(0,N-M):
    mod_dct[i] = audio_m_dct[i]
print(len(mod_dct))
```

```python
audio_m = idct(mod_dct, norm = 'ortho')
```
